```python
import json
from textblob import TextBlob  # Make sure to install the TextBlob library: pip install textblob

def analyze_sentiment(paragraph):
    # Use TextBlob for sentiment analysis
    blob = TextBlob(paragraph)
    sentiment = "positive" if blob.sentiment.polarity > 0 else "negative" if blob.sentiment.polarity < 0 else "neutral"
    return sentiment

def process_article(article):
    article_id = article["id"]
    content = article["content"]
    labels = article["labels"]
    topics = article.get("topics", {})

    # Initialize sentiments for each political typology
    typology_sentiments = {typology: {} for typology in labels}

    # Define keywords and their associated sentiments for each political typology
    keyword_sentiments = {
        "social_justice_reform": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "economic_freedom": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "inflation": {"most_extreme_right": "negative", "hyper_partisan_right": "negative", "strong_right": "negative", "skews_right": "negative"},
        "healthcare": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "climate": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "jobs_economy": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "immigration": {"most_extreme_right": "negative", "hyper_partisan_right": "negative", "strong_right": "negative", "skews_right": "negative"},
        "taxes": {"most_extreme_right": "negative", "hyper_partisan_right": "negative", "strong_right": "negative", "skews_right": "negative"},
        "government_spending": {"most_extreme_right": "negative", "hyper_partisan_right": "negative", "strong_right": "negative", "skews_right": "negative"},
        "abortion": {"most_extreme_left": "negative", "hyper_partisan_left": "negative", "strong_left": "negative", "skews_left": "negative"},
        "civil_rights": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "guns": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "national_security": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "education": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "crime": {"most_extreme_right": "negative", "hyper_partisan_right": "negative", "strong_right": "negative", "skews_right": "negative"},
        "civil_liberties": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "foreign_policy": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "criminal_justice_reform": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "democrat": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "gop": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "republican": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "socialism": {"most_extreme_left": "positive", "hyper_partisan_left": "positive", "strong_left": "positive", "skews_left": "positive"},
        "war": {"most_extreme_right": "positive", "hyper_partisan_right": "positive", "strong_right": "positive", "skews_right": "positive"},
        "police": {"most_extreme_left": "negative", "hyper_partisan_left": "negative", "strong_left": "negative", "skews_left": "negative"},
    }

    for topic, keywords_info in topics.items():
        paragraphs = keywords_info.get("paragraphs", [])

        for paragraph_info in paragraphs:
            paragraph_content = paragraph_info["content"]
            paragraph_sentiment = analyze_sentiment(paragraph_content)

            # Associate sentiment with each political typology based on keywords
            for keyword, typology_sentiments_map in keyword_sentiments.items():
                if keyword.lower() in paragraph_content.lower():
                    for typology, sentiment in typology_sentiments_map.items():
                        typology_sentiments[typology][keyword] = sentiment

    # Add sentiments to the article
    article["typology_sentiments"] = typology_sentiments

    return article

def main(input_json_path, output_json_path):
    with open(input_json_path, "r", encoding="utf-8") as json_file:
        data = json.load(json_file)

    processed_data = [process_article(article) for article in data]

    with open(output_json_path, "w", encoding="utf-8") as output_json_file:
        json.dump(processed_data, output_json_file, ensure_ascii=False, indent=2)

if __name__ == "__main__":
    input_json_path = "/kaggle/input/typology-jsn/typology_all.json"  # Replace with your input JSON file path
    output_json_path = "/kaggle/working/output.json"  # Replace with your desired output JSON file path

    main(input_json_path, output_json_path)
```
